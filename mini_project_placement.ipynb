```python
import csv
import os

USER_FILE = "users.txt"
DATA_FILE = "progress.txt"

# Create files if they don't exist
if not os.path.exists(USER_FILE):
    with open(USER_FILE, "w") as f:
        f.write("admin,admin123\n")
        f.write("nazma,python123\n")

if not os.path.exists(DATA_FILE):
    open(DATA_FILE, "w").close()

# ---------------- LOGIN ----------------
print("====== LOGIN ======")

username = input("Username: ")
password = input("Password: ")

valid = False

with open(USER_FILE, "r") as f:
    for line in f:
        u, p = line.strip().split(",")
        if username == u and password == p:
            valid = True
            break

if not valid:
    print("Invalid Username or Password")
    raise SystemExit

print("\nLogin Successful!")

# ---------------- MENU ----------------

while True:

    print("\n========== Placement Preparation Tracker ==========")
    print("1. Add Progress")
    print("2. View Progress")
    print("3. Search by Date")
    print("4. Edit Record")
    print("5. Delete Record")
    print("6. Total Problems Solved")
    print("7. Weekly Report")
    print("8. Export CSV")
    print("9. Exit")

    choice = input("Enter Choice: ")

    # Add
    if choice == "1":

        date = input("Date (DD-MM-YYYY): ")
        language = input("Language: ")
        problems = input("Problems Solved: ")
        aptitude = input("Aptitude Score: ")
        hours = input("Study Hours: ")

        with open(DATA_FILE, "a") as f:
            f.write(f"{date},{language},{problems},{aptitude},{hours}\n")

        print("Progress Saved!")

    # View
    elif choice == "2":

        if os.path.getsize(DATA_FILE) == 0:
            print("No Records Found.")
        else:
            print("\nDate\t\tLanguage\tProblems\tAptitude\tHours")
            print("-"*65)

            with open(DATA_FILE, "r") as f:
                for line in f:
                    d = line.strip().split(",")
                    print(f"{d[0]}\t{d[1]}\t\t{d[2]}\t\t{d[3]}\t\t{d[4]}")

    # Search
    elif choice == "3":

        search_date = input("Enter Date: ")

        found = False

        with open(DATA_FILE, "r") as f:
            for line in f:
                if line.startswith(search_date):
                    print(line.strip())
                    found = True

        if not found:
            print("Record Not Found.")

    # Edit
    elif choice == "4":

        edit_date = input("Enter Date to Edit: ")

        lines = []

        found = False

        with open(DATA_FILE, "r") as f:
            lines = f.readlines()

        with open(DATA_FILE, "w") as f:

            for line in lines:

                data = line.strip().split(",")

                if data[0] == edit_date:

                    language = input("New Language: ")
                    problems = input("New Problems: ")
                    aptitude = input("New Aptitude Score: ")
                    hours = input("New Study Hours: ")

                    f.write(f"{edit_date},{language},{problems},{aptitude},{hours}\n")

                    found = True

                else:
                    f.write(line)

        if found:
            print("Record Updated.")
        else:
            print("Record Not Found.")

    # Delete
    elif choice == "5":

        delete_date = input("Enter Date to Delete: ")

        lines = []

        deleted = False

        with open(DATA_FILE, "r") as f:
            lines = f.readlines()

        with open(DATA_FILE, "w") as f:

            for line in lines:

                if not line.startswith(delete_date):
                    f.write(line)
                else:
                    deleted = True

        if deleted:
            print("Record Deleted.")
        else:
            print("Record Not Found.")

    # Total
    elif choice == "6":

        total = 0

        with open(DATA_FILE, "r") as f:
            for line in f:
                if line.strip():
                    total += int(line.strip().split(",")[2])

        print("Total Problems Solved:", total)

    # Weekly Report
    elif choice == "7":

        problems = 0
        hours = 0

        with open(DATA_FILE, "r") as f:
            for line in f:
                if line.strip():
                    data = line.strip().split(",")
                    problems += int(data[2])
                    hours += float(data[4])

        print("\n===== Weekly Report =====")
        print("Total Problems :", problems)
        print("Study Hours    :", hours)

    # Export CSV
    elif choice == "8":

        with open(DATA_FILE, "r") as txt, open("progress.csv", "w", newline="") as csvfile:

            writer = csv.writer(csvfile)

            writer.writerow(["Date","Language","Problems","Aptitude","Hours"])

            for line in txt:
                if line.strip():
                    writer.writerow(line.strip().split(","))

        print("CSV Exported Successfully!")

    # Exit
    elif choice == "9":

        print("Thank You!")
        break

    else:
        print("Invalid Choice")
```

```python
import csv
import os

filename = "progress.txt"

while True:
    print("\n====== Placement Preparation Tracker ======")
    print("1. Add Progress")
    print("2. View Progress")
    print("3. Search by Date")
    print("4. Total Problems Solved")
    print("5. Weekly Summary")
    print("6. Export to CSV")
    print("7. Exit")

    choice = input("Enter your choice: ")

    # Add Progress
    if choice == "1":
        date = input("Enter Date (DD-MM-YYYY): ")
        language = input("Language: ")
        problems = int(input("Problems Solved: "))
        aptitude = int(input("Aptitude Score: "))
        hours = float(input("Study Hours: "))

        with open(filename, "a") as f:
            f.write(f"{date},{language},{problems},{aptitude},{hours}\n")

        print("Progress Added Successfully!")

    # View Progress
    elif choice == "2":

        if not os.path.exists(filename):
            print("No records found.")
            continue

        print("\nDate\t\tLanguage\tProblems\tAptitude\tHours")
        print("-" * 65)

        with open(filename, "r") as f:
            for line in f:
                data = line.strip().split(",")
                print(f"{data[0]}\t{data[1]}\t\t{data[2]}\t\t{data[3]}\t\t{data[4]}")

    # Search by Date
    elif choice == "3":

        search_date = input("Enter Date: ")
        found = False

        if os.path.exists(filename):
            with open(filename, "r") as f:
                for line in f:
                    data = line.strip().split(",")

                    if data[0] == search_date:
                        print("\nRecord Found")
                        print("Date :", data[0])
                        print("Language :", data[1])
                        print("Problems :", data[2])
                        print("Aptitude :", data[3])
                        print("Hours :", data[4])
                        found = True

        if not found:
            print("No Record Found.")

    # Total Problems
    elif choice == "4":

        total = 0

        if os.path.exists(filename):
            with open(filename, "r") as f:
                for line in f:
                    data = line.strip().split(",")
                    total += int(data[2])

        print("Total Problems Solved:", total)

    # Weekly Summary
    elif choice == "5":

        total_problems = 0
        total_hours = 0

        if os.path.exists(filename):
            with open(filename, "r") as f:
                for line in f:
                    data = line.strip().split(",")
                    total_problems += int(data[2])
                    total_hours += float(data[4])

        print("\n===== Weekly Summary =====")
        print("Problems Solved :", total_problems)
        print("Study Hours     :", total_hours)

    # Export CSV
    elif choice == "6":

        if not os.path.exists(filename):
            print("No records available.")
            continue

        with open(filename, "r") as txt, open("progress.csv", "w", newline="") as csvfile:

            writer = csv.writer(csvfile)
            writer.writerow(["Date", "Language", "Problems", "Aptitude", "Hours"])

            for line in txt:
                writer.writerow(line.strip().split(","))

        print("Exported to progress.csv")

    # Exit
    elif choice == "7":
        print("Thank You!")
        break

    else:
        print("Invalid Choice")
```
